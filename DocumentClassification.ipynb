```python
import numpy as np
import pandas as pd
```

```python
from pathlib import Path
import os.path
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
# Create a list with the filepaths for training and testing
dir_ = Path('/train')
train_filepaths = list(dir_.glob(r'**/*.jpg'))

dir_ = Path('/test')
test_filepaths = list(dir_.glob(r'**/*.jpg'))
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
# Preprocessing the Training set
train_datagen = ImageDataGenerator(rescale = 1./255,
                                   shear_range = 0.2,
                                   zoom_range = 0.2,
                                   rotation_range=20,
                                   horizontal_flip = True)


training_set=train_datagen.flow_from_directory(
    directory=r"E:/ImageData/training",
    target_size=(224, 224),
    color_mode="rgb",
    batch_size=10,
    class_mode="binary",
    shuffle=True,
    seed=0)


test_datagen = ImageDataGenerator(rescale = 1./255,validation_split=0.1)
test_set1 = test_datagen.flow_from_directory('E:/ImageData/testing',
                                            target_size=(224,224),
                                            batch_size=10,
                                            class_mode='binary',
                                            subset='validation', 
                                            seed=0,
                                            shuffle=False)
```

```python
input_shape=training_set.image_shape
input_shape
```

```python
## Model 
from tensorflow.keras import layers,Sequential
import tensorflow as tf
```

```python
## Mobile net model 
IMG_SHAPE = (224,224,3)
base_model = tf.keras.applications.MobileNetV2(input_shape=IMG_SHAPE,
                                               include_top=False, 
                                               weights='imagenet')
base_model.trainable = False
model = tf.keras.Sequential([
  base_model,
  tf.keras.layers.Conv2D(128, 3, activation='relu'),
  tf.keras.layers.Dropout(0.2),
  tf.keras.layers.GlobalAveragePooling2D(),
  tf.keras.layers.Dense(1, activation='sigmoid')])

adam = tf.keras.optimizers.Adam(lr=0.001, beta_1=0.9, beta_2=0.999, epsilon=None, decay=0.0, amsgrad=False)
model.compile(optimizer=adam,loss='binary_crossentropy',metrics=['accuracy'])
```

```python
model.fit_generator(training_set,epochs =2000,validation_data=test_set1,verbose=2)
```

```python
loss,accuracy=model.evaluate(test_set1)
print("accuracy=",accuracy)
print("loss="loss)
```

```python
model.save('/dogcatModel/document.h5')
```

```python
from tensorflow.keras.models import load_model
modelx = load_model('/dogcatModel/document.h5')
```
